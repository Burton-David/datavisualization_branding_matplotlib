```python
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rcParams
print(matplotlib.__version__)
%matplotlib inline
```

```python
matplotlib.rc_file(
    '/Library/Frameworks/Python.framework/Versions/3.8/lib/python3.8/site-packages/matplotlib/mpl-data/matplotlibrc',
    use_default_template=True)
# Data for plotting
t = np.arange(0.0, 2.4, 0.23)
s = 1 + np.sin(2* np.pi * t)

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='Negative to Positive Opinion with 1 being Neutral',
       title='My Opinion Of The Protaganist During Film Viewing')
ax.grid()

fig.savefig("test.png")
plt.show()
```
